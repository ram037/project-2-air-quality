# Import Air Quality Index CSV Files for Data Cleanup
Create connection to csv files and load all data into a pandas dataframe. Drop unnecessary columns and any null data values.

```python
aqi2018_df = pd.read_csv("Data/annual_aqi_by_county_2018.csv") #1056 rows
aqi2017_df = pd.read_csv("Data/annual_aqi_by_county_2017.csv") #1062 rows
aqi2016_df = pd.read_csv("Data/annual_aqi_by_county_2016.csv") #1054 rows
aqi2015_df = pd.read_csv("Data/annual_aqi_by_county_2015.csv") #1061 rows
aqi2014_df = pd.read_csv("Data/annual_aqi_by_county_2014.csv") #1054 rows
aqi2013_df = pd.read_csv("Data/annual_aqi_by_county_2013.csv") #1062 rows
```

```python
frames = [aqi2018_df, aqi2017_df, aqi2016_df, aqi2015_df, aqi2014_df, aqi2013_df]
joined_aqi_df= pd.concat(frames)
joined_aqi_df.head()
#6439 rows
```

```python
joined_aqi_df = joined_aqi_df.rename(columns={"State" : "state",
                                       "County" : "county",
                                       "Year" : "year",
                                       "Days with AQI" : "days_with_aqi",
                                       "Good Days" : "good_days",
                                       "Moderate Days" : "moderate_days",
                                       "Unhealthy for Sensitive Groups Days" : "unhealthy_sensitive_days",
                                       "Unhealthy Days" : "unhealthy_days",
                                       "Very Unhealthy Days" : "very_unhealthy_days",
                                       "Hazardous Days" : "hazardous_days",
                                       "Max AQI" : "max_aqi",
                                       "90th Percentile AQI" : "90th_pct_aqi",
                                       "Median AQI" : "median_aqi",
                                       "Days CO" : "days_co",
                                       "Days NO2" : "days_no2",
                                       "Days Ozone" : "days_ozone",
                                       "Days SO2" : "days_so2",
                                       "Days PM2.5" : "days_pm2.5",
                                       "Days PM10" : "days_pm10",
                                       })
morethan100 = joined_aqi_df["days_with_aqi"] > 100
aqi_df = joined_aqi_df[morethan100]
aqi_df = aqi_df.drop("days_co", 1)
aqi_df = aqi_df.drop("days_no2", 1)
aqi_df = aqi_df.drop("days_ozone", 1)
aqi_df = aqi_df.drop("days_so2", 1)
aqi_df = aqi_df.drop("days_pm2.5", 1)
aqi_df = aqi_df.drop("days_pm10", 1)
```

# Import Air Monitor CSV files for Data Cleanup

```python
mon2018 = pd.read_csv("Data/annual_conc_by_monitor_2018.csv")
mon2017 = pd.read_csv("Data/annual_conc_by_monitor_2017.csv")
mon2016 = pd.read_csv("Data/annual_conc_by_monitor_2016.csv")
mon2015 = pd.read_csv("Data/annual_conc_by_monitor_2015.csv")
mon2014 = pd.read_csv("Data/annual_conc_by_monitor_2014.csv")
mon2013 = pd.read_csv("Data/annual_conc_by_monitor_2013.csv")
```

```python
frames = [mon2018, mon2017, mon2016, mon2015, mon2014, mon2013]
air_mon = pd.concat(frames)
air_mon.head()
```

```python
ozone_df = air_mon.loc[air_mon['Parameter Name'] == "Ozone"]
ozone_df = ozone_df.loc[ozone_df['Sample Duration'] == "8-HR RUN AVG BEGIN HOUR"]
ozone_df_filt = ozone_df[[                      
                          'Parameter Name',
                          'Year',
                          'Units of Measure',
                          'Arithmetic Mean',
                          'Local Site Name',
                          'State Name',
                          'County Name'
                         ]]
```

```python
ozone_df_filt = ozone_df_filt.rename(columns={"Parameter Name" : "parameter_name",
                                              "Year" : "year",
                                              "Units of Measure" : "units_of_measure",
                                              "Arithmetic Mean" : "arithmetic_mean",
                                              "Local Site Name" : "local_site_name",
                                              "State Name" : "state",
                                              "County Name" : "county"})
ozone_df_filt.head()
```

# Import FIPS county codes and append to existing DataFrames

```python
fips_df = pd.read_excel("Data/FIPS_excel.xlsx", dtype=str)
```

```python
fips_df = fips_df.rename(columns = {"Name" : "county",
                                    "State" : "state_abbr",
                                    "State Name" : "state",
                                   "FIPS" : "fips"})
fips_df.head()
```

```python
fips_df["state_county"] = fips_df["state"].map(str) + "_" + fips_df["county"]
```

```python
fips_df.head()
```

# Map state and county to existing DataFrames

```python
aqi_df["state_county"] = aqi_df["state"].map(str) + "_" + aqi_df["county"]
aqi_df.head()
```

```python
ozone_df_filt["state_county"] = ozone_df_filt["state"].map(str) + "_" + ozone_df_filt["county"]
ozone_df_filt.head()
```

# Join FIPS code to all existing DataFrames

```python
fips_ozone_df = pd.merge(ozone_df_filt, fips_df, on="state_county")
fips_ozone_df = fips_ozone_df.rename(columns={"state_x" : "state",
                                                "county_x" : "county"})
fips_ozone_df = fips_ozone_df.drop("county_y", 1)
fips_ozone_df = fips_ozone_df.drop("state_y", 1)
fips_ozone_df.head()
```

```python
# Each county for each year has more than one measurement. We need to take the average of all those measurements.
final_ozone_df = fips_ozone_df.groupby(['state_county',
                                        'fips',
                                        'parameter_name',
                                        'year'
                                       ])['arithmetic_mean'].mean().reset_index()
```

```python
final_ozone_df
```

```python
final_aqi_df = pd.merge(aqi_df, fips_df, on="state_county")
final_aqi_df = final_aqi_df.rename(columns={"state_x" : "state",
                                                "county_x" : "county"})
final_aqi_df = final_aqi_df.drop("county_y", 1)
final_aqi_df = final_aqi_df.drop("state_y", 1)
goodperc = final_aqi_df["good_days"]/final_aqi_df["days_with_aqi"]*100
modperc = final_aqi_df["moderate_days"]/final_aqi_df["days_with_aqi"]*100
unsentperc = final_aqi_df["unhealthy_sensitive_days"]/final_aqi_df["days_with_aqi"]*100
unhealthperc = final_aqi_df["unhealthy_days"]/final_aqi_df["days_with_aqi"]*100
vunhealthperc = final_aqi_df["very_unhealthy_days"]/final_aqi_df["days_with_aqi"]*100
hazperc = final_aqi_df["hazardous_days"]/final_aqi_df["days_with_aqi"]*100
final_aqi_df["good_percentage"] = goodperc
final_aqi_df["moderate_percentage"] = modperc
final_aqi_df["unhealthy_sensitive_percentage"] = unsentperc
final_aqi_df["unhealthy_percentage"] = unhealthperc
final_aqi_df["very_unhealthy_percentage"] = vunhealthperc
final_aqi_df["hazardous_percentage"] = hazperc
final_aqi_df.head()
```

```python
aqi_2018 = final_aqi_df.loc[final_aqi_df["year"] == 2018, :]
goodperc = aqi_2018["good_days"]/aqi_2018["days_with_aqi"]
modperc = aqi_2018["moderate_days"]/aqi_2018["days_with_aqi"]
unsentperc = aqi_2018["unhealthy_sensitive_days"]/aqi_2018["days_with_aqi"]
unhealthperc = aqi_2018["unhealthy_days"]/aqi_2018["days_with_aqi"]
vunhealthperc = aqi_2018["very_unhealthy_days"]/aqi_2018["days_with_aqi"]
hazperc = aqi_2018["hazardous_days"]/aqi_2018["days_with_aqi"]
aqi_2018["good_percentage"] = goodperc
aqi_2018["moderate_percentage"] = modperc
aqi_2018["unhealthy_sensitive_percentage"] = unsentperc
aqi_2018["unhealthy_percentage"] = unhealthperc
aqi_2018["very_unhealthy_percentage"] = vunhealthperc
aqi_2018["hazardous_percentage"] = hazperc
aqi_2018 = aqi_2018.sort_values(by=["good_percentage"])
aqi_2018.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    y= aqi_2018["state_county"],
    x= aqi_2018["good_percentage"],
    name='Good Days',
    orientation='h',
    marker=dict(
        color='rgb(25, 100, 126)')
    )
)
fig.add_trace(go.Bar(
    y= aqi_2018["state_county"],
    x= aqi_2018["moderate_percentage"],
    name='Moderate Days',
    orientation='h',
    marker=dict(
        color='rgb(17,157,164)')
    )
)
fig.add_trace(go.Bar(
    y= aqi_2018["state_county"],
    x= aqi_2018["unhealthy_sensitive_percentage"],
    name='Unhealthy Sensitive Days',
    orientation='h',
    marker=dict(
        color='rgb(255,177,36)')
    )
)
fig.add_trace(go.Bar(
    y= aqi_2018["state_county"],
    x= aqi_2018["unhealthy_percentage"],
    name='Unhealthy Days',
    orientation='h',
    marker=dict(
        color='rgb(75,63,114)')
    )
)
fig.add_trace(go.Bar(
    y= aqi_2018["state_county"],
    x= aqi_2018["very_unhealthy_percentage"],
    name='Very Unhealthy Days',
    orientation='h',
    marker=dict(
        color='rgb(198,63,114)')
    )
)
fig.add_trace(go.Bar(
    y= aqi_2018["state_county"],
    x= aqi_2018["hazardous_percentage"],
    name='Hazardous Days',
    orientation='h',
    marker=dict(
        color='rgb(207,68,65)')
    )
)
# fig.update_yaxes(range = [0,1000])
fig.update_layout(title_text = "Air Quality Proportions by County",
                  barmode='stack',
                  height = 15000,
                  width = 1000)
fig.show()
```
