## Importing CSVs and merging

```python
mon2018 = pd.read_csv("C:/Users/brian/Project2_AirDataQuality/Data/annual_conc_by_monitor_2018.csv")
mon2017 = pd.read_csv("C:/Users/brian/Project2_AirDataQuality/Data/annual_conc_by_monitor_2017.csv")
mon2016 = pd.read_csv("C:/Users/brian/Project2_AirDataQuality/Data/annual_conc_by_monitor_2016.csv")
mon2015 = pd.read_csv("C:/Users/brian/Project2_AirDataQuality/Data/annual_conc_by_monitor_2015.csv")
mon2014 = pd.read_csv("C:/Users/brian/Project2_AirDataQuality/Data/annual_conc_by_monitor_2014.csv")
mon2013 = pd.read_csv("C:/Users/brian/Project2_AirDataQuality/Data/annual_conc_by_monitor_2013.csv")
```

```python
frames = [mon2018, mon2017, mon2016, mon2015, mon2014, mon2013]
```

## Cleaning and filtering data

```python
ozone_df = air_mon.loc[air_mon['Parameter Name'] == "Ozone"]
ozone_df = ozone_df.loc[ozone_df['Sample Duration'] == "8-HR RUN AVG BEGIN HOUR"]
```

```python
ozone_df_filt = ozone_df[['State Code',
                          'County Code',                        
                          'Parameter Name',
                          'Year',
                          'Units of Measure',
                          'Arithmetic Mean',
                          'Local Site Name',
                          'State Name',
                          'County Name'
                         ]]
```

```python
ozone_df_filt
```

```python
fips_df = pd.read_excel("C:/Users/brian/Project2_AirDataQuality/Data/FIPS Excel.xlsx", dtype=str)
```

```python
fips_df
```

```python
fips_df["state_county"] = fips_df["State Name"].map(str) + "_" + fips_df["Name"]
```

```python
fips_df
```

```python
ozone_df_filt
```

```python
ozone_df_filt["state_county"] = ozone_df_filt["State Name"].map(str) + "_" + ozone_df_filt["County Name"]
```

```python
ozone_df_filt
```

```python
fips_df_filt = fips_df.drop(["Name", "State", "State Name"], axis=1)
```

```python
fips_df_filt
```

```python
df_merge = pd.merge(ozone_df_filt, fips_df_filt, on="state_county")
```

```python
df_merge
```

```python
df_merge = df_merge.drop(["state_county"], axis=1)
```

```python
df_merge
```

```python
df_groupby = df_merge.groupby(['FIPS', 'Year'], as_index=False).mean()
```
